```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
from PIL import Image
```

```python
df = pd.read_excel('Книга1.xlsx')
```

```python
text = " ".join(df['item_name'])
text
```

```python
cloud = WordCloud().generate(text = text)
cloud
```

```python
figure_heinz = np.array(Image.open('Heinz-symbol.jpg'))
```

```python
cloud = WordCloud(mask=figure_heinz, colormap='Set3').generate(text = text)

fig,ax = plt.subplots(figsize = (16,15))
ax.imshow(cloud)
ax.axis('off')
plt.show()
```
