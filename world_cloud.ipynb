```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_excel('Лист Microsoft Excel.xlsx')
```

```python
text = " ".join(df['SKU_обработан_1.1.545'])
text
```

```python
cloud = WordCloud().generate(text = text)
cloud
```

```python
fig,ax = plt.subplots(figsize = (16,15))
ax.imshow(cloud)
ax.axis('off')
plt.show()
```
