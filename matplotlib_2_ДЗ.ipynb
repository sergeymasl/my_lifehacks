```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%config IPCompleter.greedy=True
```

### Загрузка данных

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объема продаж всех игр жанра "Sports" в Японии. Задайте графику размер 12 на 6

```python
data_Japan = df[df['Genre'] == 'Sports'].groupby('Year').agg('sum')['JP_Sales']

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(data_Japan)
ax.set_xlabel('Год')
ax.set_ylabel('Объем продаж, млн.копий')
ax.set_title('Динамика объема продаж игр жанра Spotrts в Японии')

plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
fig, ax = plt.subplots(figsize = (12, 6))

for region in (['Global_Sales','NA_Sales', 'EU_Sales', 'JP_Sales']):
    data_Activision = df[df['Publisher'] == 'Activision'].groupby('Year').agg('sum')[region]
    ax.plot(data_Activision, label = str(region))
    
ax.set_xlabel('Год')
ax.set_ylabel('Объем продаж, млн.копий')
ax.set_title('Динамика продаж студии Activision в Северной Америке, Европе, Японии и всем мире')
ax.legend()
plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
fig, axs = plt.subplots(nrows = 4,figsize = (12, 12))

for i, region  in enumerate(['Global_Sales','NA_Sales', 'EU_Sales', 'JP_Sales']):
    data_Activision = df[df['Publisher'] == 'Activision'].groupby('Year').agg('sum')[region]
    axs[i].plot(data_Activision, label = region)
    axs[i].set_ylabel('Объем продаж, млн.копий')
    axs[i].legend()
    if i == 3:
        axs[i].set_xlabel('Год')
fig.suptitle('Динамики продаж студии "Activision"', y = 0.9) 
plt.show()
```

### Задача 4

Постройте такие же линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире на одном полотне в сетке 2 на 2 графика. Итоговое полотно сделайте размером 12 на 12. Все графики должны разделять оси X и Y

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 12), sharey = True)
max_y = df[df['Publisher'] == 'Activision'].groupby('Year').agg('sum')['Global_Sales']

for i, region  in enumerate(['Global_Sales','NA_Sales', 'EU_Sales', 'JP_Sales']):
    data_Activision = df[df['Publisher'] == 'Activision'].groupby('Year').agg('sum')[region]
    r = 0 if i == 0 or i == 1 else 1
    c = 0 if i == 0 or i == 2 else 1
    axs[r][c].plot(data_Activision, label = region)
    if c == 0:
        axs[r][c].set_ylabel('Объем продаж, млн.копий')
    if r == 1:
        axs[r][c].set_xlabel('Год')
    axs[r][c].grid(color = 'lightgrey', linestyle = '--')
    axs[r][c].legend()
fig.suptitle('Динамики продаж студии "Activision"', y = 0.9)    
plt.show()
```

### Задача 5

В одной координатной сетке постройте 2 гистограммы распределения мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010 года (включая 2010 год). Графики сделайте полупрозрачными, отсечки должны быть общими и соответствовать ширине столбцов обоих графиков

```python
fig, ax = plt.subplots(figsize = (12, 6))

data_Micro = df[(df['Publisher'] == 'Microsoft Game Studios') & (df['Year'] >= 2010)]['Global_Sales']
data_Take = df[(df['Publisher'] == 'Take-Two Interactive') & (df['Year'] >= 2010)]['Global_Sales']

n, bins, pathes =ax.hist(data_Take, label = 'Take-Two Interactive', alpha = 0.5, bins = 10)
ax.hist(data_Micro, label = 'Microsoft Game Studios', alpha = 0.5, bins = bins)
ax.set_ylabel('# записей')
ax.set_xlabel('Объем продаж, млн.копий')
ax.set_xticks(bins)
ax.grid(color = 'lightgrey', linestyle = '--')
ax.set_title('Hаспределения мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010 года')
ax.legend()
    
plt.show()
```

### Задача 6

Определите 5 платформ, для которых в Японии было продано больше всего игр за все время. Проиллюстрируйте величину продаж на столбчатой диаграмме, столбец с самым высоким значением окрасьте зеленым, а с самым низким - красным

```python
data_platform = pd.DataFrame(df.groupby('Platform').agg('sum')['JP_Sales'].sort_values(ascending=False).head(5))
data_platform['Color'] = '#598234'
data_platform.loc[data_platform['JP_Sales'] == data_platform['JP_Sales'].max(),'Color'] = '#78A5A3'
data_platform.loc[data_platform['JP_Sales'] == data_platform['JP_Sales'].min(),'Color'] = '#CE5A57'
fig, ax = plt.subplots(figsize = (12, 6))

ax.bar(data_platform.index, data_platform['JP_Sales'], color = data_platform['Color'], alpha = 1, width=0.4)
ax.set_ylabel('Объем общемировых продаж, млн.копий')
ax.set_xlabel('Платформа')
ax.set_title('Топ 5 платформ по продажам в Японии')


plt.show()
```

### Задача 7

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зеленого полупрозрачного прямоугольника

```python
data_nin = pd.DataFrame(df[df['Publisher'] == 'Nintendo'].groupby('Year').agg('count')['Rank'])
data_nin_35 = data_nin[data_nin['Rank'] >= 35]
year_max = data_nin_35[data_nin_35['Rank'] == data_nin_35['Rank'].max()].index[0]
year_min = data_nin_35[data_nin_35['Rank'] == data_nin_35['Rank'].min()].index[0]

fig, ax = plt.subplots(figsize = (12, 6))

ax.plot(data_nin, color = '#F25C00')
ax.set_ylabel('Количество выпущенных игр в год')
ax.set_xlabel('Год')
ax.set_title('Динамика выпуска игр компанией Nintendo')
ax.axvspan(xmin = year_min, xmax = year_max, color = '#258039', alpha = 0.5,\
           label = 'Период с продажими более 35 игр в год')
ax.grid(color = 'lightgrey', linestyle = '--')

plt.xticks(data_nin.index, rotation = 45)
ax.legend()

plt.show()
```

### Задача 8

Определите 3 жанра и 4 издателя с самыми большими продажами игр в мире за все время. Постройте сетку графиков 4 на 3, в каждой ячейке постройте точечный график, проверяющий зависимость общемировых продаж от продаж в Северной Америке для каждой пары (жанр, издатель). Размер итогового полотна определите самостоятельно таким образом, чтобы все графики были видны

```python
top_Genre = df.groupby(['Genre']).agg('sum')['Global_Sales'].sort_values(ascending=False).head(3)
top_Publisher = df.groupby(['Publisher']).agg('sum')['Global_Sales'].sort_values(ascending=False).head(4)
fig, axs = plt.subplots(nrows = 4, ncols = 3, figsize = (16,16))
for row in range(4):
    for colum in range(3):
        Publisher = top_Publisher[top_Publisher == top_Publisher[row]].index[0]
        Genre = top_Genre[top_Genre == top_Genre[colum]].index[0]
        data_top = df[(df['Publisher'] == Publisher) & (df['Genre'] == Genre)].groupby('Year').agg('sum')
        axs[row][colum].scatter(data_top['Global_Sales'], data_top['NA_Sales'])
        axs[row][colum].grid(color = 'lightgrey', linestyle = '--')
        if row == 0:
            axs[row][colum].set_title('Жанр: ' + Genre)
        if colum == 0:
            axs[row][colum].set_ylabel('Издатель: ' + Publisher + '\n\nОбъем продаж в Северной Америке\nмлн копий')
        if row == 3:
            axs[row][colum].set_xlabel('Объем общемировых продаж, млн копий')
fig.suptitle('Сравнение общемировых продаж и продаж в Северной Америке в топ 3 жанрах у топ 4 издателей', y = 0.93)
plt.show()
```
